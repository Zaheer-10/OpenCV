Thresholding is a technique of image processing that converts an image into a binary image, where the pixels are either 0 or 255. Thresholding can be useful for segmenting foreground and background objects, enhancing edges, removing noise, etc.

There are different types of thresholding methods in OpenCV, such as **simple thresholding**, **Otsuâ€™s thresholding**, and **adaptive thresholding**.

### What is Adaptive Thresholding?
    Adaptive is a way of making an image black and white that changes depending on how bright or dark the image is. Unlike simple, where we use one number to decide if a pixel is black or white, adaptive looks at a small area around each pixel and uses a different number for that area.

    This way, adaptive can work better for images that have different brightness in different parts, like shadows, lights, etc., where one number may not work well for the whole image.



- "Adaptive thresholding is a method of thresholding that adapts to the local variations in the image brightness. 
- Unlike simple thresholding, where a global threshold value is applied to all pixels, adaptive thresholding considers a small neighborhood of pixels around each pixel and computes a local threshold value for that region."

## How to Apply Adaptive Thresholding 

In OpenCV, the `cv2.adaptiveThreshold` function is used to perform adaptive thresholding. 
- It has the following syntax:
#### cv2.adaptiveThreshold(src, max_val, adaptive_method, threshold_type, block_size, constant)
---
The parameters are:

- src: The source image, which should be a grayscale image.
- max_val: The maximum value to assign to the pixels that pass the threshold test.
- adaptive_method: The method to calculate the local threshold value. It can be one of the following:
    - cv2.ADAPTIVE_THRESH_MEAN_C: The threshold value is the mean of the neighborhood area minus the constant C.
    - cv2.ADAPTIVE_THRESH_GAUSSIAN_C: The threshold value is the weighted sum of the neighborhood area minus the constant C. The weights are a Gaussian window.
- threshold_type: The type of thresholding to apply. It can be one of the following:
- cv2.THRESH_BINARY: If the pixel value is greater than the threshold value, it is set to max_val. Otherwise, it is set to 0.
- cv2.THRESH_BINARY_INV: If the pixel value is greater than the threshold value, it is set to 0. Otherwise, it is set to max_val.
- block_size: The size of the neighborhood area to calculate the local - - threshold value. It must be an odd number greater than 1.
- constant: The constant C to subtract from the mean or weighted mean.

---
The function returns the thresholded image as a NumPy array.

## Example

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('Images/sudoko.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

ret1, thresh1 = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

ret2, thresh2 = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

thresh3 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                cv2.THRESH_BINARY, 11, 5)
thresh4 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                cv2.THRESH_BINARY, 11, 5)

titles = ["Original Image", "Simple", "Otsu's",
          "Adaptive Mean", "Adaptive Gaussian"]
images = [gray, thresh1, thresh2, thresh3, thresh4]

for i in range(5):
    plt.subplot(2, 3, i + 1)
    plt.imshow(images[i], "gray")
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])

plt.show()
```

## Simple threshold: 
- This method uses one global threshold value for the whole image. 
- If the pixel value is smaller than the threshold, it is set to 0 (black), otherwise it is set to 255 (white). 
- This method is easy to apply but may not work well for images with uneven lighting or noise.

## Otsuâ€™s threshold: 
- This method automatically finds the optimal threshold value that maximizes the difference between the two classes of pixels (foreground and background). 
- It assumes that the image has a bimodal histogram (two peaks). 
- This method can work better than simple threshold for images with a clear separation of foreground and background.

## Adaptive mean: 
- This method calculates the threshold value for each pixel based on the mean of its local neighborhood (a small region around it). 
- It subtracts a constant C from the mean to get the threshold. 
- This method can adapt to the local variations in brightness and contrast in the image.

## Adaptive Gaussian: 
- This method is similar to adaptive mean, but instead of using the mean, it uses the weighted average of the local neighborhood, where the weights are based on a Gaussian distribution (a bell-shaped curve). 
- This method can give a smoother result than adaptive mean.

## Application: 
- Adaptive thresholding can be used for separating things in images that have different lighting or noise, such as papers, car plates, x-rays, etc.

---
## Advantages: 
- Adaptive thresholding can do better than using one threshold value for the whole image. 
- It can also save time and effort from finding the best threshold value.
---
### Disadvantages: 
- Adaptive thresholding can be slower and harder than using one threshold value. 
- It can also make some parts of the image worse, depending on what property and size of the area we use.
