```python
import cv2 as cv
import numpy as np 
import matplotlib.pyplot as pyplot 

img = cv.imread('Images\messi1.jpg')
img = cv.cvtColor(img , cv.COLOR_BGR2RGB)

GRADIENTS_SOBELx = cv.Sobel(img ,-1 , 1 , 0 )
GRADIENTS_SOBELy = cv.Sobel(img ,-1 , 0 , 1 )

GRADIENTS_SOBELxy = cv.addWeighted(GRADIENTS_SOBELx , 0.5 , GRADIENTS_SOBELy , 0.5 , 0)

GRADIENT_LAPLACIAN = cv.Laplacian(img , -1 )

CANNY_OUTPUT = cv.Canny(img , 80 , 150 )

cv.imshow('SOBELx ‚ûó' , GRADIENTS_SOBELx)
cv.imshow('SOBELy üòÜ' , GRADIENTS_SOBELy)
cv.imshow('LAPLACIAN üëâ' , GRADIENT_LAPLACIAN)
cv.imshow('CANNY ‚ù§Ô∏è‚Äçüî•' ,CANNY_OUTPUT )
cv.waitKey(0)
```
